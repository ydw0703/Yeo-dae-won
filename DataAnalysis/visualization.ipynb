```python
#형태소 분석 패키지
from konlpy.tag import Twitter
#단어 카운팅 패키지
from collections import Counter
import pandas as pd
import openpyxl
import csv
```

```python
file = open("C:/Coding/comment_youtube.txt",'r','utf-8') #'r'파읽을 읽기전용으로 함.
lists = file.readlines()
file.close()
lists
```

```python
twitter = Twitter()
morphs = []
for sentence in lists:
    morphs.append(twitter.pos(sentence))
    
print(morphs)
```

```python
noun_adj_adv_list=[]

for sentence in morphs :
    for word, tag in sentence :
        if tag in ['Noun'] and ("것" not in word) and ("내" not in word)and ("나" not in word)and ("수"not in word) and("명"not in word)and("말"not in word)and("왜"not in word):
            noun_adj_adv_list.append(word)
            
print(noun_adj_adv_list)
```

```python
count = Counter(noun_adj_adv_list)
words = dict(count.most_common())
```
