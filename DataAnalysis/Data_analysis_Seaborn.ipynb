```python
#importing libraries and dataset
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Read dataset
df = pd.read_csv('C:/Users/Administrator/jupyter/stationery.csv',index_col=0)
df.head()
```

```python
#선형 회귀 시각화
sns.lmplot(x='unit',y='price_per_unit',data=df,hue='name')
plt.ylim(0,None)
```

```python
# 상자 그림 그리기 (모든 숫자의 상자그림을 그린다.)
sns.boxplot(data=df)
```

```python
# 필요없는 값들을 제거
stats_df = df.drop(['unit_stock'],axis=1)
sns.boxplot(data=stats_df)
```

```python
# Swarm plot with Pokemon color palette
sns.swarmplot(x='unit', y='totla_price', data=df)
```

```python
# 그래프를 밑에 그림과 같이 출력해보자. bins는 갯수지정. 숫자만 가능하다.
sns.distplot( df['unit'], bins = 15 )
```

```python
# 모든 표를 나타낸다. 'sex'를 또하나의 구분기준으로 만들어서.
sns.pairplot(df,hue = 'name')
```

```python
# 막대 그래프에 각 축을 만들어, 값들을 df테이블의 값들
sns.barplot(x='place', y = 'unit', data = df)
```

```python
# 막대 그래프 테이블의 특정 값의 개수를 세어주는 함수.
sns.countplot(x='place', data = df)
```

```python
tc=df.corr()
sns.heatmap(tc, annot = True)
```

```python
# df을 정렬하자. 'fare'값을 기준으로 내림차순으로. 'fare'값을 기준 열로 만들자. .cumsum을 활용하여 누적합을 나타내자.
titanic_ds.sort_values('fare',ascending=False)\
.reset_index()['fare']\
.cumsum()\
.plot()
```
